```python
#Retirada de colunas que já sei que irão atrapalhar o modelo futuramente
base.drop("ID", axis = 1, inplace= True)
base.drop("Z_Revenue", axis = 1, inplace= True)
base.drop("Z_CostContact", axis = 1, inplace= True)
base.head()
```

```python
#Tratamento de valores nulos na coluna "Income" pela mediana dos valores da coluna
base["Income"] = base["Income"].fillna((base["Income"].median()))
base.isnull().sum()
```

```python
# Como pretendo utilizar árvore de decisão como modelo, já vou simplificar um pouco o banco de dados, somando todas as compras do usuário e criar uma coluna de faixas etárias e simplificar a data de
# entrada do usuário

base["Purchases"] = base["NumDealsPurchases"] + base["NumWebPurchases"] + base["NumCatalogPurchases"] + base["NumStorePurchases"]
base.drop("NumDealsPurchases", axis = 1, inplace= True)
base.drop("NumWebPurchases", axis = 1, inplace= True)
base.drop("NumCatalogPurchases", axis = 1, inplace= True)
base.drop("NumStorePurchases", axis = 1, inplace= True)
base.head()
```

```python
#2020 é a data que o banco de dados foi postado
base["Age"] = 2020 - base["Year_Birth"]
base.head()
```

```python
base.drop("Year_Birth", axis = 1, inplace= True)
base.head()
```

```python
base.drop(base.loc[base["Age"] > 80]. index, inplace = True)
base.describe()
```

```python
#Criar coluna de Faixa Etária
idade_bins = [0,30,40,50,60,80]
idade_categoria = ["Até 30 anos", "De 31 a 40", "De 41 a 50", "De 51 a 60", "Acima de 60"]
base["Faixa Etária"] = pd.cut(base["Age"], idade_bins, labels = idade_categoria)
base["Faixa Etária"].value_counts()
```

```python
base.drop("Dt_Customer", axis = 1, inplace= True)
```

```python
#Transformação dos dados Object em numéricos
lb = LabelEncoder()
base['Education'] = lb.fit_transform(base['Education'])
base['Marital_Status'] = lb.fit_transform(base['Marital_Status'])
base['Faixa Etária'] = lb.fit_transform(base["Faixa Etária"])

base.dropna(inplace = True)
```

```python
#Separação do target do restante dos dados
base.info()
```

```python
target = base.iloc[:,19:20]
target.info()
```

```python
preditoras = base.copy()
del preditoras["Response"]
preditoras.head()
```

```python
#Divisão em dados de Treino e Teste
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(preditoras, target, test_size= 0.3)
```

```python
#Treinamento do Modelo de Machine Learning
modelo = DecisionTreeClassifier()
modelo.fit(X_treinamento, y_treinamento)
```

```python
#Criação da variável de previsões do modelo
previsoes = modelo.predict(X_teste)
previsoes
```

```python
#Averiguação da Acurácia do Modelo
taxa_acerto = accuracy_score(y_teste, previsoes)
taxa_acerto
```
